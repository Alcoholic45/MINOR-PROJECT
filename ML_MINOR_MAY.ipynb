```python
import os
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.transform import resize

target =[]
images = []
flat_data = []

DATADIR = '/content/images'  #from where i have to consider the images 
CATEGORIES = ['mango','apple']

for category in CATEGORIES:  #print(category)
  class_num = CATEGORIES.index(category) #print(class_num) we r converting the mango and apple in 0 and 1 by label encoding
  path = os.path.join(DATADIR,category) #create path to use all images use print(path)
  for img in os.listdir(path):#we r listing the images from path
     img_array = imread(os.path.join(path,img)) #here we r reading some of the images from the path and mentioning the img to show
     #print(img_array) 
     #print(img_array.shape) shows the all the image shapes
     plt.imshow(img_array)
     img_resized = resize(img_array,(150,150,3)) # resizing
     flat_data.append(img_resized.flatten()) #flattenig
     images.append(img_resized) # append means adding we are adding resized images to image
     target.append(class_num) # we are adding the class_num to target

#converting into numpy
flat_data = np.array(flat_data)
images = np.array(images)
target = np.array(target)


```

```python
from skimage.color import rgb2gray

img_resize = resize(img_array,(50,50))
img_gray = rgb2gray(img_resize)
plt.imshow(img_gray,cmap='gray')
img_gray.size
```

```python
#splitting
from sklearn.model_selection import train_test_split # we have to test the data whether the accuaracy is good or bad 
x_train,x_test,y_train,y_test = train_test_split(flat_data,target,test_size=0.3,random_state =42)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm
param_grid = [
              {'C':[1,10,100,1000],'kernel':['linear']},
              {'C':[1,101,100,1000],'gamma':[0.001,0.0001],'kernel':['rbf']}
]
svc = svm.SVC(probability=True)
grid = GridSearchCV(svc,param_grid)
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(x_test)
y_pred
```

```python
y_test
```

```python
#evaluation
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```
